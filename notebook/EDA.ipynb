```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/Sample.csv", encoding="latin1")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])
```

```python
df.dtypes
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
df.drop(["Row ID","Order ID","Customer ID","Customer Name","Product ID",   "Product Name"], axis=1, inplace=True)
```

```python
df.head()
```

```python
print("Total Sales:", df["Sales"].sum())
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Sales"], bins=20)
plt.title("Sales Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
df.groupby("Category")["Sales"].sum().plot(kind="bar")
plt.title("Sales by Category")
plt.show()
```

```python
plt.figure(figsize=(8,5))
df.groupby("Region")["Sales"].sum().plot(kind="bar")
plt.title("Sales by Region")
plt.show()
```

```python
df.head()
```

```python
df["year"] = df["Order Date"].dt.year
df["month"] = df["Order Date"].dt.month
df["day"] = df["Order Date"].dt.day
df["dayofweek"] = df["Order Date"].dt.dayofweek
df["quarter"] = df["Order Date"].dt.quarter
```

```python
df["shipping_days"] = (df["Ship Date"] - df["Order Date"]).dt.days
df["profit_margin"] = df["Profit"] / df["Sales"]
```

```python
df["discount_amount"] = df["Sales"] * df["Discount"]
```

```python
daily_sales = df.groupby("Order Date")["Sales"].sum().reset_index()

print(daily_sales.head())
```

```python
daily_sales["lag_1"] = daily_sales["Sales"].shift(1)
daily_sales["lag_7"] = daily_sales["Sales"].shift(7)

daily_sales["rolling_mean"] = daily_sales["Sales"].rolling(7).mean()

daily_sales.dropna(inplace=True)
```

```python
daily_sales["year"] = daily_sales["Order Date"].dt.year
daily_sales["month"] = daily_sales["Order Date"].dt.month
daily_sales["day"] = daily_sales["Order Date"].dt.day
daily_sales["dayofweek"] = daily_sales["Order Date"].dt.dayofweek
```

```python
daily_sales.head()
```

```python
daily_sales.to_csv("data/daily_sales.csv", index=False)
```
